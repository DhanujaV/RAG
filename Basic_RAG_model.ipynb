```python
from sentence_transformers import SentenceTransformer
import faiss
import openai
```

```python
docs=[
    "Large Language Models (LLMs) are artificial intelligence systems trained on massive datasets to generate human-like text and understand context. Retrieval-Augmented Generation (RAG) is a technique that expands these models' capabilities by fetching specific facts from external knowledge bases.",
    "Electrical transformers are passive devices that transfer electrical energy between circuits without changing the frequency, utilizing electromagnetic induction",
    "Figma is a cloud-based digital design and prototyping tool used primarily for UI/UX (User Interface and User Experience) design. It allows teams to brainstorm, design, build, and test applications, websites, and digital products collaboratively and in real-time right from a web browser.",
    "Streamlit is an open-source Python library that allows developers, data scientists, and ML engineers to build interactive, shareable web applications in pure Python. It eliminates the need for traditional front-end development, meaning you can create beautiful dashboards and AI/ML tools without writing HTML, CSS, or JavaScript.",
    "Natural Language Processing (NLP) is a branch of Artificial Intelligence (AI) that gives machines the ability to read, understand, and interpret human language. It bridges the gap between human communication and computer understanding by breaking down speech and text into data that algorithms can comprehend and respond to."

]
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings=model.encode(docs).astype("float32")
```

```python
doc_embeddings.shape
```

```python
dimension=doc_embeddings.shape[1]
index=faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
query="how figma works?"
query_embedding=model.encode([query]).astype("float32")
top_k=2
distances,indices=index.search(query_embedding,top_k)
retrieved_chunks=[docs[i] for i in indices[0]]
print("retrieved chunks")
for chunk in retrieved_chunks:
  print("-",chunk)
```

```python
from groq import Groq

client = Groq(
    api_key="************************************************"
)
```

```python
context="\n".join(retrieved_chunks)
prompt=f"""
Answer the question based only on the context below.
Context:
{context}
Question:
{query}
Answer:
"""
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role":"user","content":prompt}
    ],
    temperature=0.2
)
print("Finsl answer:\n")
print(response.choices[0].message.content)


```
